```python
from torchvision.datasets import ImageFolder
from torchvision import transforms
from torch.utils.data import Dataset, DataLoader
```

if we dont use a transform, when we use the dataloader we'll get the following error: <br>
TypeError: default_collate: batch must contain tensors, numpy arrays, numbers, dicts or lists; found <class 'PIL.Image.Image'>


so let's create a transform and convert the PILimage to tensors

```python
transform = transforms.Compose([transforms.Grayscale(),
                                transforms.Resize((150,150)),
                                transforms.ToTensor()
                                ])
```

```python
traindataset = ImageFolder(root='data/train/', transform=transform)
```

### train a model

Let's create a model and train it just for fun. I dont expect it to do well

```python
import torch
from torch import nn
import torch.nn.functional as F
from torch import optim
```

```python
class Model1(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(1, 10, 3)
        self.pool = nn.MaxPool2d(2)
        self.conv2 = nn.Conv2d(10, 10, 3)
        self.fc1 = nn.Linear(10*36*36, 10)
        
    def forward(self, x):
        x = F.relu(self.pool(self.conv1(x)))
        x = F.relu(self.pool(self.conv2(x)))
        x = torch.flatten(x, 1) # used instead of x = x.view(x.size(0), -1)
        x = torch.log_softmax(self.fc1(x), dim=1)
        
        return x
        
```

```python
model1 = Model1()
model1
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model1.parameters())
```

```python
len(traindataset)
```

```python
# hyperparameters
epochs = 1
batch_size = 50
```

```python
trainloader = DataLoader(traindataset, batch_size, shuffle=True)
```

```python
print_every = 100 # print every x batches
running_loss = []
for i,(images, labels) in enumerate(trainloader):
    optimizer.zero_grad()
    outputs = model1(images)
    loss = criterion(outputs, labels)
    loss.backward()
    optimizer.step()
    
    running_loss.append(loss.item())    
    if i%print_every == 0:
        print(f'iteration: {i}, loss: {loss.item()}')
```

```python
plt.plot(np.arange(len(running_loss)), running_loss)
plt.xlabel('iteration')
plt.ylabel('loss')
```
