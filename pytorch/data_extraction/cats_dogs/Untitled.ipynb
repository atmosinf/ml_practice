```python
import pandas as pd
import os
```

```python
catlist = os.listdir('cats/')
```

```python
catlist
```

```python
catconcat = [[i,0] for i in catlist]
catconcat
```

```python
doglist = os.listdir('dogs/')
doglist
```

```python
dogconcat = [[i,1] for i in doglist]
dogconcat
```

```python
fulllist = catconcat + dogconcat
fulllist
```

```python
len(fulllist)
```

```python
df = pd.DataFrame(fulllist, columns=['image', 'label'])
```

```python
df
```

```python
df.to_csv('cat_dog_metadata.csv', sep=',', index=False)
```
