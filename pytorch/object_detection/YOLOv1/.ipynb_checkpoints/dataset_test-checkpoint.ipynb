in the above output, the first 20 elements show which class the bbox belongs to, the 20th element shows if there is an object or not, and the 21 - 24  elements show the dimensions of the bbox relative to the cell size (by default there are 7 cells)

```python
plt.figure(figsize=(10,7))
for i,row in enumerate(label_mat[:,:,20]):
    for j,cell in enumerate(row):
        if cell == 1:
            print('i,j ->', i,j) # which grid cell contains the image
            print('label_mat[i,j] ->', label_mat[i,j])
            
            imagec = np.ascontiguousarray(image, dtype=np.float64)
#             midpoint and dimensions relative to the grid cell, as returned by the custom dataset
            midpoint_x, midpoint_y = label_mat[i,j,21:23].numpy()
            width, height = label_mat[i,j,23:25].numpy()
            print('midpoint and dimensions relative to the grid cell')
            print(midpoint_x, midpoint_y)
            print(width, height)
            
            cell_width = int(image.shape[1] / 7)
            cell_height = int(image.shape[0] / 7)
            
#            convert midpoint and dimensions to be relative to the image so that it can be visualized using cv.rectangle 
            midpoint_x_image, midpoint_y_image = (j+midpoint_x) * cell_width, (i + midpoint_y) * cell_height
            width_image, height_image = width * cell_width, height * cell_height
            print('midpoint and dimensions relative to the image')
            print(midpoint_x_image, midpoint_y_image)
            print(width_image, height_image)
            
            cvrect = cv.rectangle(imagec, (int(midpoint_x_image - width_image/2), int(midpoint_y_image - height_image/2)), \
                                  (int(midpoint_x_image + width_image/2), int(midpoint_y_image + height_image/2)), (255,255,255))
            plt.imshow(imagec)
```
