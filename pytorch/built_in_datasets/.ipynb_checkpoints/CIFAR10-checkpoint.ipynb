### in this notebook we'll try to load and use a built-in pytorch dataset 

```python
import torch
import torchvision
from torchvision import transforms
import matplotlib.pyplot as plt
import numpy as np
```

```python
from torch.utils.data import DataLoader
```

```python
transform = transforms.Compose([transforms.ToTensor(),
                               transforms.Normalize((0.5,0.5,0.5), (0.5,0.5,0.5))])
```

```python
root = 'CIFAR10'
dataset = torchvision.datasets.CIFAR10(root, train=True, transform=transform, download=True)
```

```python
batch_size = 5
classes = ['plane', 'car', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
```

```python
trainloader = DataLoader(dataset, batch_size, True)
```

```python
# get some random training images
dataiter = iter(trainloader)
images, labels = dataiter.next()

# show images
imshow(torchvision.utils.make_grid(images))
# print labels
print(' '.join(f'{classes[labels[j]]:7s}' for j in range(batch_size)))
```

### split the dataset into train and test sets

```python
from torch.utils.data import random_split
```

```python
trainlen = int(0.7 * len(dataset))
testlen = int(0.15 * len(dataset))
vallen = int(len(dataset) - (trainlen + testlen))
```

```python
trainlen, testlen, vallen
```

```python
traindataset, testdataset, valdataset = random_split(dataset, [trainlen, testlen, vallen])
```

```python
len(traindataset), len(testdataset), len(valdataset)
```

### create the model 

```python
import torch.nn as nn
import torch.nn.functional as F


class Net(nn.Module):
    def __init__(self):
        super().__init__()
        self.conv1 = nn.Conv2d(3, 6, 5)
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(6, 16, 5)
        self.fc1 = nn.Linear(16 * 5 * 5, 120)
        self.fc2 = nn.Linear(120, 84)
        self.fc3 = nn.Linear(84, 10)

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = torch.flatten(x, 1) # flatten all dimensions except batch
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x
```

```python
net = Net()
```

```python
net
```

```python
from torch import optim
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(net.parameters(), lr=0.001, momentum=0.9)
```

```python
epochs = 2
```

```python
# test tqdm
from tqdm import tqdm
```

```python
for e in range(epochs):  # loop over the dataset multiple times

    running_loss = 0.0
    loop = tqdm(enumerate(trainloader), total=len(trainloader), leave=True)
    for i, data in loop:
        # get the inputs; data is a list of [inputs, labels]
        inputs, labels = data

        # zero the parameter gradients
        optimizer.zero_grad()

        # forward + backward + optimize
        outputs = net(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        # print statistics
        running_loss += loss.item()
#         if i % 2000 == 1999:    # print every 2000 mini-batches
#             print(f'[{epoch + 1}, {i + 1:5d}] loss: {running_loss / 2000:.3f}')
#             running_loss = 0.0
        
        # update progress bar
        loop.set_description(f'epoch: [{e+1}/{epochs}]')
        loop.set_postfix(loss = loss.item(), acc = torch.rand(1).item())

print('Finished Training')
```
