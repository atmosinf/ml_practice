```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('ggplot')
```

```python
x = np.random.randn(10)
y = np.random.randn(10)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
plt.subplot(131)
plt.xlabel('1')
plt.title('1')
plt.scatter(x,y)
plt.subplot(132)
plt.xlabel('2')
plt.title('2')
plt.scatter(x,y)
plt.subplot(133)
plt.xlabel('3')
plt.title('3')
plt.scatter(x,y)
plt.scatter(y,x) # just testing
plt.tight_layout()
```

```python
fig, ax = plt.subplots(1,3)
plt.tight_layout()
ax[0].set_title('1')
ax[0].scatter(x, y)
ax[1].set_title('2')
ax[2].set_title('3')
```

```python
fig1 = plt.figure()
fig1.add_subplot(131)
plt.plot([1,2,3],[1,2,3])
fig1.add_subplot(132)
plt.plot([1,2,3],[1,2,3])
fig1.add_subplot(333)
plt.plot([1,2],[2,1])
fig1.add_subplot(336)
plt.plot([1,2],[2,1])
fig1.add_subplot(339)
plt.plot([1,2],[2,1])
plt.tight_layout()
fig2 = plt.figure()
fig2.add_subplot(311)
plt.plot([1,2],[1,2])
plt.tight_layout()
```

```python
plt.subplot(131)
plt.plot([1,2],[1,2])
plt.subplot(132)
plt.plot([1,2],[1,2])
plt.subplot(333)
plt.plot([1,2],[1,2])
plt.subplot(336)
plt.plot([1,2],[1,2])
plt.subplot(339)
plt.plot([1,2],[1,2])
plt.tight_layout()
```

```python
fig = plt.figure()

# Adds a subplot at the 1st position
fig.add_subplot(221)
plt.plot([1, 2, 3], [1, 2, 3])

# Adds a subplot at the 4th position
fig.add_subplot(2, 2, 4)
plt.plot([3, 2, 1], [1, 2, 3])

fig.show()
```
