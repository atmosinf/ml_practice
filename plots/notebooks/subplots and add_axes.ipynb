### in this notebook we'll test the differences and working of .subplot, .subplots, .add_subplot, .add_axes

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('ggplot')
```

generate some random data

```python
x = np.random.randn(10)
y = np.random.randn(10)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
```

test .subplot

```python
plt.subplot(131)
plt.xlabel('1')
plt.title('1')
plt.scatter(x,y)
plt.subplot(132)
plt.xlabel('2')
plt.title('2')
plt.scatter(x,y)
plt.subplot(133)
plt.xlabel('3')
plt.title('3')
plt.scatter(x,y)
plt.scatter(y,x) # just testing
plt.tight_layout()
```

subplots using ax

```python
fig, ax = plt.subplots(1,3)
plt.tight_layout()
ax[0].set_title('1')
ax[0].scatter(x, y)
ax[1].set_title('2')
ax[2].set_title('3')
```

test fig.add_subplot

```python
fig1 = plt.figure()
fig1.suptitle('main title 1')
fig1.add_subplot(131)
plt.plot([1,2,3],[1,2,3])
fig1.add_subplot(132)
plt.plot([1,2,3],[1,2,3])
fig1.add_subplot(333)
plt.plot([1,2],[2,1])
fig1.add_subplot(336)
plt.plot([1,2],[2,1])
fig1.add_subplot(339)
plt.plot([1,2],[2,1])
plt.tight_layout()
fig2 = plt.figure()
fig2.suptitle('main title 2')
fig2.add_subplot(311)
plt.plot([1,2],[1,2])
plt.tight_layout()
```

and again using .subplot

```python
plt.suptitle('main title')
plt.subplot(131)
plt.plot([1,2],[1,2])
plt.title('subtitle')
plt.subplot(132)
plt.plot([1,2],[1,2])
plt.subplot(333)
plt.plot([1,2],[1,2])
plt.subplot(336)
plt.plot([1,2],[1,2])
plt.subplot(339)
plt.plot([1,2],[1,2])
plt.tight_layout()
```

### inference

as far as i see, the main difference between .subplot and .add_subplot is that .add_subplot is applied on fig and not plt, which means that you can have multiple figures in a plot. the example above shows that (fig1 and fig2)

### let's now test add_axes

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
```

```python
fig = plt.figure()
fig.subplots_adjust(top = 0.8)
ax1 = fig.add_subplot(211)
  
t = np.arange(0.0, 1.0, 0.01)
s = np.sin(2 * np.pi * t)
line, = ax1.plot(t, s, color ='green', lw = 2)
  
np.random.seed(19680801)
  
ax2 = fig.add_axes([0.15, 0.1, 0.7, 0.3])
n, bins, patches = ax2.hist(np.random.randn(1000), 50,
                            facecolor ='green',
                            edgecolor ='yellow')
   
fig.suptitle('matplotlib.figure.Figure.add_axes() \
function Example\n\n', fontweight ="bold")
  
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.0, 0.0, 0.5, 0.5]) # x,y,w,h
ax1.set_title('1')
ax2 = fig.add_axes([0.0, -0.5, 0.5, 0.5])
ax2.set_title('2')
plt.tight_layout()
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.0, 0.0, 0.5, 0.5]) # x,y,w,h
ax1.set_title('1')
ax1.plot([1,2],[1,2])
ax2 = fig.add_axes([0.6, 0.0, 1, 1])
ax2.set_title('2')
ax2.plot([1,2],[1,2])
plt.tight_layout()
```

```python
fig = plt.figure()

for i in range(5):
    ax = fig.add_axes([i,i,0.8,0.8])
    ax.plot([1,2],[1,2])
    ax.set_title(i)
```

```python
fig = plt.figure()

for i in range(5):
    ax = fig.add_axes([0,i,0.8,0.8])
    ax.plot([1,2],[1,2])
    ax.set_title(i)
```

```python
fig = plt.figure()

for i in range(5):
    ax = fig.add_axes([i,0,0.8,0.8])
    ax.plot([1,2],[1,2])
    ax.set_title(i)
```
